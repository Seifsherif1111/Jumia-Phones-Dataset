```python
from bs4 import BeautifulSoup as bs
import requests
import re
```

# Important tags:
        'a' attributes : <br>
        brand :'data-ga4-item_brand'<br>
        id : 'data-ga4-item_id'<br>
        name : 'data-ga4-item_name'<br>
        product-link : href<br><br>
        (div class= "img-c")
        -> img tag -> <br>
        img_link: 'data-src'<br><br>
        (div class ="info")-> price :(div class = "prc")<br>
        (div class="s-prc-w") ->old price :(div class = "old")<br>
        discount percentage:(div class="bdg _dsct _sm")<br>
        number of reviews:(div class = "rev") -> rating:(div class= "stars_s")<br>
        
        


```python
header = {'User-Agent':'PhoneScraperBot (+mailto:seif.sherif.ai@gmail.com)'} #added email and clearly identified myself as a bot
products = []
for i in range(1,50): #under limit per minute allowed by robot.txt
    response = requests.get(f'https://www.jumia.com.eg/phones-tablets/?q=phones&operating_system=Android--iOS&page={i}#catalog-listing',headers=header)
    html= bs(response.text,'html.parser')
    items= html.find_all('a',class_ = 'core')
    for item in items:
        record = {
            'phone-brand':item.get('data-ga4-item_brand'),
            'Phone-id':item.get('data-ga4-item_id'),
            'phone-name':item.get('data-ga4-item_name'),
            'priceEGP': item.find('div',class_="prc").text.split()[1],
            'old-priceEGP': item.find('div',class_="old").text.split()[1] if item.find('div',class_="old") !=None else item.find('div',class_="prc").text.split()[1] ,
            'rating-out-of-5': item.find('div',class_="rev").text.split()[0] if item.find('div',class_="rev")!=None else 0 ,
            'reviews-count':re.split(r'[()]',item.find('div',class_="rev").text.split()[-1])[1] if item.find('div',class_="rev")!=None else 0 ,
            'discount': item.find('div',class_="bdg _dsct _sm").text if item.find('div',class_="bdg _dsct _sm") !=None else '0%',
            'product-link': "https://www.jumia.com.eg"+ item.get('href'),
            'image-link':item.find('img').get('data-src')
        }
        products.append(record)
```

```python
import pandas as pd
df = pd.DataFrame(products)
df
```

```python
df.to_csv("Jumia phones.csv")
```
